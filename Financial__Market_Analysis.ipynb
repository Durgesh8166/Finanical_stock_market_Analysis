```python
daily_df=pd.read_csv('/content/AAPL_daily_update.csv')
daily_df.head()
```

```python
daily_df.isna().sum()
```

```python
daily_df.info()
```

```python
#convert Data from object->type  to datatime type
daily_df['Date']=pd.to_datetime(daily_df['Date'])
```

```python
daily_df.set_index('Date', inplace=True)
```

```python
daily_df
```

```python
daily_df['Daily_Return']=daily_df['Adj Close'].pct_change()
```

```python
daily_df['Daily_Return'].fillna(0)
daily_df.head()
```

```python
daily_df['MA_15'] = daily_df['Adj Close'].rolling(window=15).mean()
daily_df
```

```python
daily_df['MA_15']
```

```python
daily_df['MA_30'] = daily_df['Adj Close'].rolling(window=30).mean()
daily_df.tail(30)
```

```python
daily_df['Volatility'] =daily_df['Daily_Return'].rolling(window=10).std()
daily_df.head(22)
```

```python
#Calculate 12-day EMA
daily_df['EMA_12'] = daily_df['Adj Close'].ewm(span=12, adjust=False).mean()
daily_df.head(12)
```

```python
daily_df['EMA_26'] = daily_df['Adj Close'].ewm(span=26, adjust=False).mean()
daily_df
```

```python
plt.plot(daily_df.index,daily_df['Adj Close'],label='Adj close')
plt.title('AAPL Adjusted Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price (USD)')
plt.grid(True)  # Add gridlines for better readability
plt.legend()
plt.show()
```

```python


# Assuming your DataFrame is named 'df' and contains the specified columns
# Also assuming the Date column is converted to datetime and set as index.

def plot_combined_graph(daily_df):
    """Plots Adj Close, EMA12, EMA26, Volatility, MA_50, and MA_200 on a single graph."""

    plt.figure(figsize=(15, 8))  # Adjust figure size for better readability

    plt.plot(daily_df.index, daily_df['Adj Close'], label='Adj Close', linewidth=2)
    plt.plot(daily_df.index, daily_df['EMA_12'], label='EMA 12', linestyle='--')
    plt.plot(daily_df.index, daily_df['EMA_26'], label='EMA 26', linestyle='--')
    plt.plot(daily_df.index, daily_df['Volatility'], label='Volatility', alpha=0.7)  # Use alpha for transparency
    plt.plot(daily_df.index, daily_df['MA_15'], label='MA 15', linestyle=':')
    plt.plot(daily_df.index, daily_df['MA_30'], label='MA 30', linestyle=':')

    plt.title('AAPL Stock Analysis')
    plt.xlabel('Date')
    plt.ylabel('Price / Volatility')
    plt.grid(True)
    plt.legend(loc='best')  # Let Matplotlib choose the best legend location
    plt.show()

# Example usage (assuming your DataFrame is named 'df'):
plot_combined_graph(daily_df)
```

```python

def plot_daily_returns_distribution(daily_df):
    """Plots the distribution of daily returns (histogram and KDE)."""

    plt.figure(figsize=(12, 6))

    # Histogram
    sns.histplot(daily_df['Daily_Return'], bins=50, kde=False, label='Histogram')

    # KDE (Kernel Density Estimate)
    sns.kdeplot(daily_df['Daily_Return'], label='KDE')

    plt.title('AAPL Daily Returns Distribution')
    plt.xlabel('Daily Return')
    plt.ylabel('Frequency / Density')
    plt.legend()
    plt.grid(True)
    plt.show()


# Example usage (assuming your DataFrame is named 'daily_df'):
# Assuming your DataFrame is named 'daily_df' and contains the specified columns
# Also assuming the Date column is converted to datetime and set as index.

#Example data, replace with your 'daily_df' data.

plot_daily_returns_distribution(daily_df)
```
